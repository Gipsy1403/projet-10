```python
# Importation des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

# Importation des fichiers
esn = pd.read_csv("df_merge.csv")

pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)

esn.head(3)
```

```python
# suppression des colonnes poste et departement et celles qui n'ont qu'une valeur
esn_clean=esn.drop(columns=["poste","departement","nombre_heures_travailless","nombre_employee_sous_responsabilite", "ayant_enfants" ])
esn_clean.info()
```

```python
# valeurs uniques pour toutes les colonnes du dataframe
for col in esn_clean.columns:
    print(f"Valeurs uniques pour {col}:")
    print(esn_clean[col].unique())
    print() 
```

```python
# Creation d'une nouvelle feature : moyenne des scores de satisfaction.
def add_global_satisfaction(esn_clean):
    esn_clean = esn_clean.copy()
    
    satisfaction_cols = [
        "satisfaction_employee_environnement",
        "satisfaction_employee_nature_travail",
        "satisfaction_employee_equipe",
        "satisfaction_employee_equilibre_pro_perso"
    ]
    
    esn_clean["score_satisfaction_global"]=esn_clean[satisfaction_cols].mean(axis=1)
    
    return esn_clean


# Creation d'une nouvelle feature : indice d'evolution dans l'entreprise.
def add_evolution_index(esn_clean):
    esn_clean = esn_clean.copy()
    
    esn_clean["indice_evolution"] = (
        esn_clean["annees_dans_l_entreprise"]
        - esn_clean["annees_depuis_la_derniere_promotion"]
        - esn_clean["augementation_salaire_precedente"]
    )
    
    return esn_clean

# fonction globale de feature engineering
def feature_engineering(esn_clean):
    esn_clean = add_global_satisfaction(esn_clean)
    esn_clean = add_evolution_index(esn_clean)
    
    return esn_clean

df_fe = feature_engineering(esn_clean)
df_fe.head()

df_fe.groupby("a_quitte_l_entreprise")[
    ["score_satisfaction_global", "indice_evolution"]
].mean()

# résultat : Les employés restés dans l’entreprise ont en moyenne une satisfaction plus élevée et un indice d’évolution plus important que ceux qui sont partis. Cela suggère que la satisfaction et la progression dans l’entreprise sont liées au fait de rester ou quitter.
```

```python
# Transformation de la colonne "a quitte l'entreprise" en nombre afin de définir la cible y
def prepare_target(esn_clean, target_col="a_quitte_l_entreprise"):
	y=esn_clean[target_col].apply(lambda x:1 if x =="oui" else 0)
	return y

y=prepare_target(df_fe)


print(y.value_counts())
```

```python
# Retirer la colonne id et la colonne cible y afin de sélectionner les features (X)
def select_features(esn_clean, target_col="a_quitte_l_entreprise", drop_cols=None):
	if drop_cols is None:
		drop_cols=["id",target_col]
		X=esn_clean.drop(columns=drop_cols)
		return X

X=select_features(df_fe)
X.head(3)
```

```python
# Encodage des variables qualitatives
def encode_features(X):
	X_encoded=X.copy()

	# encodage des colonnes ordinales
	statut={'célibataire': 0, 'marié(e)': 1, 'divorcé(e)': 2}
	X_encoded["statut_marital"]=X_encoded["statut_marital"].map(statut)
	frequency_move={"aucun":0, "occasionnel":1, "frequent":2}
	X_encoded["frequence_deplacement"]=X_encoded["frequence_deplacement"].map(frequency_move)

	# encodage des colonnes nominales
	col_nominales=X_encoded.select_dtypes(include="object").columns.tolist()

	# OneHotEncoder en evitant la multiplicité des colonnes grâce à drop_first à True
	X_encoded=pd.get_dummies(X_encoded,columns=col_nominales,drop_first=True, dtype=int)
	return X_encoded

X_encoded=encode_features(X)
X_encoded.head(3)

	
```

MODELISATION

```python
X_final=X_encoded_clean
y_final=y
```

```python
# Séparation des données pour des jeux d'apprentissage et de test

X_train, X_test, y_train, y_test = train_test_split(
    X_final,
    y_final,
    test_size=0.2,
    random_state=42,
    # mis car les classes sont déséquilibrées
    stratify=y_final
)
```

```python
# mettre X_train et X_test à la même échelle (scaling)
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
# Ajuste le scaler sur X_train et transformer X_train
X_train_scaled = scaler.fit_transform(X_train)

# Transforme X_test avec le même scaler (sans fit !)
X_test_scaled = scaler.transform(X_test)
```

```python
# MODELE DUMMY

# Vérification si le modèle vaut le coup
from sklearn.dummy import DummyClassifier
# instance du modèle avec toujours la classe la plus fréquente dans y_train
dummy_classifier= DummyClassifier(strategy="most_frequent")
dummy_classifier.fit(X_train_scaled, y_train)

y_dummy_pred = dummy_classifier.predict(X_test_scaled)

score_train=dummy_classifier.score(X_train_scaled, y_train)
score_test=dummy_classifier.score(X_test_scaled, y_test)

print("Score Dummy Train :", score_train)
print("Score Dummy Test :", score_test)

# le dummy Train regarde quelle classe est la plus fréquente dans y train, de fait, il prédit toujours cette classe.
# le dummy Test prédit 84% des employés appartenant à cette classe. Cela signifie que mon dataframe est déséquilibré
# Tout modèle de machine learning devra donc dépasser cette baseline pour être considéré comme pertinent
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_dummy_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["non", "oui"],
            yticklabels=["non", "oui"])
plt.xlabel("Prédit")
plt.ylabel("Réel")
plt.title("Matrice de confusion du DummyClassifier")
plt.show()

# résultat : modèle déséquilibré car ne prédit que la classe majoritaire qui dans le modèle est non
```

```python
# MODELE LINEAIRE

from sklearn.linear_model import LogisticRegression

model_linear = LogisticRegression(max_iter=1000)
model_linear.fit(X_train_scaled, y_train)
y_pred = model_linear.predict(X_test_scaled)

score_train = model_linear.score(X_train_scaled, y_train)
score_test = model_linear.score(X_test_scaled, y_test)

print("Score Logistic Train :", score_train)
print("Score Logistic Test :", score_test)

# résultat : Le modèle linéaire est supérieur (+2.7%) au 84% du dummy ce qui signifie que le modèle apprend quelque chose
# en comparant le train du test du modèle linaire, je vérifie le surapprentissage(overfiting). Dans ce cas, j'ai 2.5% d'écart qui reste raisonnable
# Donc le modèle générale bien, pas de gros surapprentissage et le modèle est stable cependant il faut faire attention car le dataframe est déséquilibré entre ceux qui restent et ceux qui partent
```

```python
# Pour cela, la métrique la plus importante pour ce type de projet est le Recall de la classe 1=oui soit départ
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
confusion_matrix(y_test, y_pred)

# résultat : le modèle prédit correctement 239 employés qui restent, ce qui est très bon mais il détecte seulement 16 employés (soit 34% des départs) qui partent réellement et en rate 31, ce qui fait que le modèle rate beaucoup de départs
# Quant à la précision, le modèle à raison à 67% mais rate beaucoup de départs. 
# Donc actuellement, le modèle est bon pour reconnaitre les employés qui restent et est moyen pour détecter les départs.
```

```python
# modèle avec équilibrage des classes
model_linear = LogisticRegression(max_iter=1000, class_weight="balanced")

# entrainement
model_linear.fit(X_train_scaled, y_train)

# prédictions
y_pred_balanced = model_linear.predict(X_test_scaled)

# accuracy
print("Score Train :", model_linear.score(X_train_scaled, y_train))
print("Score Test :", model_linear.score(X_test_scaled, y_test))

# rapport complet
print(classification_report(y_test, y_pred_balanced))

# matrice de confusion
print(confusion_matrix(y_test, y_pred_balanced))

# résultat : L’utilisation de class_weight="balanced" permet de corriger le déséquilibre des classes.Le modèle détecte beaucoup plus de départs (32 au lieu de 16 précédemment) mais l'accuracy baisse (78%): c'est normal et attendu
# Ce modèle détecte beaucoup plus de départs mais plus d'alertes, mais il permet d'anticiper les départs
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

#  Courbe de Précision-Rappel (RP)

# probabilités
y_scores = model_linear.predict_proba(X_test_scaled)[:,1]

# calcul precision / recall
precision, recall, thresholds = precision_recall_curve(y_test, y_scores)

# calcul AUC spécifique pour les classes déséquilibrées
auc_pr = average_precision_score(y_test, y_scores)
print("AUC Precision-Recall :", auc_pr)

# graphique
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Courbe Precision - Recall (PR)")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# création du modèle
rf_model = RandomForestClassifier(
    random_state=42,
    class_weight="balanced"
)

# entrainement
rf_model.fit(X_train_scaled, y_train)

# prédictions
y_pred_rf = rf_model.predict(X_test_scaled)

# scores
print("Score Train :", rf_model.score(X_train_scaled, y_train))
print("Score Test :", rf_model.score(X_test_scaled, y_test))

print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))

# Résultat : le score Train est parfait mais le score de Test est moins bien ce qui signifie que le modèle fait du surapprentissage (overfit). Le Recall est de 13% qui est nettement moins bien par rapport à celui de la régression logistique (68%).
# ce modèle mémorise bien les employés qui restent mais rate presque tous les départs (41).
```

```python
# nouveau modèle random forest
rf_model = RandomForestClassifier(
    n_estimators=300,              # nombre d'arbres
    max_depth=10,                  # limite la profondeur, pour forcer l'arbre à généraliser
    min_samples_leaf=5,            # minimum d'observations par feuille
    class_weight="balanced",
    random_state=42
)

# entrainement
rf_model.fit(X_train_scaled, y_train)

# prédictions
y_pred_rf = rf_model.predict(X_test_scaled)

# scores
print("Score Train :", rf_model.score(X_train_scaled, y_train))
print("Score Test :", rf_model.score(X_test_scaled, y_test))

# rapport
print(classification_report(y_test, y_pred_rf))

# matrice de confusion
print(confusion_matrix(y_test, y_pred_rf))

# résultat : Le modèle fait moins de surapprentissage que le modèle précédent. Le Recall est à 43% ce qui également mieux mais n'est pas suffisant
```

```python
# VALIDATION CROISEE avec REGRESSION LOGISTIQUE

# Est ce que mon modèle généralise bien ? pour cela je vais faire une validation croisée qui permet de garder l'équilibre des classes dans chaque fold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve

# création de liste pour stocker les scores
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

# création de la validation croisée
skf=StratifiedKFold(
	n_splits=5,
	shuffle=True,
	random_state=42
)
for train_index, test_index in skf.split(X_train_scaled, y_train):

    # 1. créer les sous ensembles
    X_fold_train = X_train_scaled[train_index]
    X_fold_test = X_train_scaled[test_index]

    y_fold_train = y_train.iloc[train_index]
    y_fold_test = y_train.iloc[test_index]

    # 2. créer le modèle
    model = LogisticRegression(
        max_iter=1000,
        random_state=42,
        class_weight="balanced"
    )

    # 3. entraîner le modèle
    model.fit(X_fold_train, y_fold_train)

    # 4. prédire
    y_pred = model.predict(X_fold_test)

    # 5. calculer les métriques
    acc = accuracy_score(y_fold_test, y_pred)
    prec = precision_score(y_fold_test, y_pred)
    rec = recall_score(y_fold_test, y_pred)
    f1 = f1_score(y_fold_test, y_pred)

    # 6. stocker les résultats
    accuracy_list.append(acc)
    precision_list.append(prec)
    recall_list.append(rec)
    f1_list.append(f1)

print("Accuracy :", np.mean(accuracy_list), "+/-", np.std(accuracy_list))
print("Precision :", np.mean(precision_list), "+/-", np.std(precision_list))
print("Recall :", np.mean(recall_list), "+/-", np.std(recall_list))
print("F1 :", np.mean(f1_list), "+/-", np.std(f1_list))

# Résultat : le modèle prédit à raison seulement ~38% des départs, ce qui laisse. Il détecte ~75% des départs ce qui est très bien d'autant plus que l'écart type (0.4%) est petit ce qui en fait un modèle stable sur les différents folds
# Le F1 est moyen car la précision est faible mais pour ce cas le recall est prioritaire, donc le F1 est acceptable
# je vais chercher le seuil afin de voir comment le modèle se comporte
```

```python
# VALIDATION CROISEE avec RANDOMFOREST

# Est ce que mon modèle généralise bien ? pour cela je vais faire une validation croisée qui permet de garder l'équilibre des classes dans chaque fold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, precision_recall_curve

# création de liste pour stocker les scores
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

# création de la validation croisée
skf=StratifiedKFold(
	n_splits=5,
	shuffle=True,
	random_state=42
)
for train_index, test_index in skf.split(X_train_scaled, y_train):

    # 1. créer les sous ensembles
    X_fold_train = X_train_scaled[train_index]
    X_fold_test = X_train_scaled[test_index]

    y_fold_train = y_train.iloc[train_index]
    y_fold_test = y_train.iloc[test_index]

    # 2. créer le modèle
    model = RandomForestClassifier(
    n_estimators=300,              # nombre d'arbres
    max_depth=10,                  # limite la profondeur, pour forcer l'arbre à généraliser
    min_samples_leaf=5,            # minimum d'observations par feuille
    class_weight="balanced",
    random_state=42
	)

    # 3. entraîner le modèle
    model.fit(X_fold_train, y_fold_train)

    # 4. prédire
    y_pred = model.predict(X_fold_test)

    # 5. calculer les métriques
    acc = accuracy_score(y_fold_test, y_pred)
    prec = precision_score(y_fold_test, y_pred)
    rec = recall_score(y_fold_test, y_pred)
    f1 = f1_score(y_fold_test, y_pred)

    # 6. stocker les résultats
    accuracy_list.append(acc)
    precision_list.append(prec)
    recall_list.append(rec)
    f1_list.append(f1)

print("Accuracy :", np.mean(accuracy_list), "+/-", np.std(accuracy_list))
print("Precision :", np.mean(precision_list), "+/-", np.std(precision_list))
print("Recall :", np.mean(recall_list), "+/-", np.std(recall_list))
print("F1 :", np.mean(f1_list), "+/-", np.std(f1_list))

# Résultat : le modèle prédit à raison seulement ~38% des départs, ce qui laisse. Il détecte ~75% des départs ce qui est très bien d'autant plus que l'écart type (0.4%) est petit ce qui en fait un modèle stable sur les différents folds
# Le F1 est moyen car la précision est faible mais pour ce cas le recall est prioritaire, donc le F1 est acceptable
# je vais chercher le seuil afin de voir comment le modèle se comporte
```

```python
# RECHERCE DU MEILLEUR SEUIL AVEC REGRESSION LOGISTIQUE

# liste de seuils à tester
thresholds = [0.3, 0.4, 0.5, 0.6]

# création de la validation croisée stratifiée
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# dictionnaire pour stocker les résultats
results = {t: {"accuracy": [], "precision": [], "recall": [], "f1": []} for t in thresholds}

# boucle sur les folds
for fold, (train_index, test_index) in enumerate(skf.split(X_train_scaled, y_train), 1):
    # créer les sous ensembles
    X_fold_train = X_train_scaled[train_index]
    X_fold_test = X_train_scaled[test_index]
    y_fold_train = y_train.iloc[train_index]
    y_fold_test = y_train.iloc[test_index]
    
    # créer et entraîner le modèle
    model = LogisticRegression(max_iter=1000, class_weight="balanced", random_state=42)
    model.fit(X_fold_train, y_fold_train)
    
    # prédictions probabilistes
    y_proba = model.predict_proba(X_fold_test)[:,1]
    
    # tester chaque seuil
    for t in thresholds:
        y_pred = (y_proba >= t).astype(int)
        results[t]["accuracy"].append(accuracy_score(y_fold_test, y_pred))
        results[t]["precision"].append(precision_score(y_fold_test, y_pred))
        results[t]["recall"].append(recall_score(y_fold_test, y_pred))
        results[t]["f1"].append(f1_score(y_fold_test, y_pred))

# afficher les résultats moyens et écart-types pour chaque seuil
for t in thresholds:
    print(f"\n=== Seuil = {t} ===")
    print("Accuracy :", np.mean(results[t]["accuracy"]), "+/-", np.std(results[t]["accuracy"]))
    print("Precision :", np.mean(results[t]["precision"]), "+/-", np.std(results[t]["precision"]))
    print("Recall :", np.mean(results[t]["recall"]), "+/-", np.std(results[t]["recall"]))
    print("F1 :", np.mean(results[t]["f1"]), "+/-", np.std(results[t]["f1"]))
```
